```python
import pandas as pd
import seaborn as sns 
```

```python
sales =pd.read_excel("C:\pythoncode\data\sales.xlsx")
```

```python
sales.head()
```

```python
sns.relplot('Sales','Profit', data=sales, col='Customer Segment',col_wrap=2)
```

```python
sns.relplot('Sales','Profit', data=sales, col='Customer Segment',col_wrap=2,hue='Ship Mode')
```

```python
sns.relplot('Sales','Profit', data=sales, col='Customer Segment',row='Product Category',hue='Ship Mode')
```

```python
sns.catplot('Ship Mode','Sales',data=sales)
```

```python
sns.catplot('Ship Mode','Sales',data=sales,jitter=False)
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit>1000'),jitter=False)
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit>1000'),kind='swarm')#swarm shows the magnitude 
```

```python
sns.catplot('Sales','Ship Mode',data=sales.query('Profit>1000'),kind='swarm')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,hue='Ship Mode',kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,row='Ship Mode',kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,col='Ship Mode',kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,col='Ship Mode',row='Customer Segment',kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,hue='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'],kind='box')
```
