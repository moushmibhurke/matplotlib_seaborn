```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal") #from ovel to circle
plt.pie(exp_vals, labels=exp_labels,shadow=True)
plt.show() #it will hide unnesserory info
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,autopct='%1.1f%%',radius=1.5, shadow=True) #shows percentage representation radius dydefault 1
plt.show() 
```

```python
#explode will explode the pie 
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0,0], counterclock=True, startangle=45)

plt.savefig("piechart1.png", bbox_inches="tight",pad_inches=2, transperant=True) #plt.show() will not allow coping picture 
```
