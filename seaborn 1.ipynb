```python
import pandas as pd
import seaborn as sns 
```

```python
sales =pd.read_excel("C:\pythoncode\data\sales.xlsx")
```

```python
sales.head()
```

```python
sns.relplot('Sales','Profit', data=sales)
```

```python
sns.relplot('Sales','Profit', data=sales, hue ='Order Priority')
```

```python
sns.relplot('Sales','Profit', data=sales, hue='Order Priority',style='Order Priority')
```

```python
sns.relplot('Sales','Profit', data=sales, size='Discount')
```

```python
sns.relplot('Sales','Profit', data=sales, size='Discount',sizes=(25,200))
```

```python
sns.relplot('Order Date','Sales',data=sales,kind='line')
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot('year','Sales',data=sales,kind='line')
```

```python
sns.relplot('year','Sales', data=sales, kind='line', ci=None) 
```

```python
#ci= confidance interval(line shows the mean vale and shadded area is +/- 95%tile 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', ci='sd') 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', estimator=None) # estimator=None shows the exact value
```

```python
sns.relplot('year','Sales', data=sales, kind='line', estimator=sum) 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', hue='Ship Mode') 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', hue='Ship Mode', style='Product Category') 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', hue='Ship Mode', style='Product Category',ci=None) 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', hue='Ship Mode', col='Product Category',ci=None) 
```

```python
sns.relplot('year','Sales', data=sales, kind='line', hue='Ship Mode', col='Region',ci=None, col_wrap=4) 
```
