```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,52,56,53,48,50,56]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5, linestyle='dotted')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b+--')
```

```python
# Same effect as 'b+' format string
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='blue',marker='+',linestyle='--')
```

```python
plt.plot(x,y,color='#888820',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1 transparancy level from 0 to 1
```
